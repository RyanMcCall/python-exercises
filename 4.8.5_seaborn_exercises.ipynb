### Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. (Choose a cutoff point for what constitutes catholic)

I will use a supermajority of 75% to represent whether the province is Catholic.

```python
swiss['is_catholic'] = swiss.Catholic >= 75
swiss
```

### Does whether or not a province is Catholic influence fertility?

```python
sns.boxplot(data=swiss, y='Fertility', x='is_catholic')
plt.xlabel("\nProvince over 75% Catholic")
plt.ylabel('Fertility\n')
plt.title("Does Being a Catholic Province Influence Fertility\n")
plt.show()
```

### What measure correlates most strongly with fertility?

```python
swiss_wo_bool = swiss[['Fertility', 'Agriculture', 'Examination', 'Education', 'Catholic',
       'Infant.Mortality']]
swiss_wo_bool
```

```python
swiss_corr = swiss.corr()
swiss_corr
```

```python
fert_corr = sns.heatmap(swiss_corr[['Fertility']].sort_values(by=['Fertility']), annot=True)
bottom, top = fert_corr.get_ylim()
fert_corr.set_ylim(bottom + 0.5, top - 0.5)
plt.title("What feature is most strongly correlated with Fertility?\n")
plt.show()
```

Fertility has the strongest correlation with Education and it is a negative correlation.

```python
sns.regplot(data=swiss, x='Fertility', y='Education')
plt.title("Does Education effect Fertility?")
plt.show()
```

Here we can see that correlation displayed again.

## Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
orders = pd.read_sql('SELECT * FROM orders', env.get_url('chipotle'))
orders
```

```python
orders.item_price = orders.item_price.str.replace('$', '').astype(float)
orders
```

```python
orders.groupby('item_name').item_price.sum().nlargest(4)
```

```python
top_4_revenue = (
    # Returns a dataframe where a column matches the check
    orders.loc[
        # This line checks if an item name from the database is in
        # the top 4 revenue producers
        orders.item_name.isin(orders.groupby('item_name').item_price.sum().nlargest(4).index)
    ]
)
top_4_revenue
```

```python
sns.barplot(data=top_4_revenue, x='item_name', y='item_price', estimator=np.sum)
plt.title('Revenue of Top 4 Items')
plt.xlabel('')
plt.ylabel('Total Revenue ($)')
```
