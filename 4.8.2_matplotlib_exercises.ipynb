## Using matplotlib to plot the following equation

$ y = x^2 - x + 2 $

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
x = list(range(25))
y = [n**2 - n + 2 for n in x]

plt.figure(figsize=(8, 4.5))
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$x^2-x+2$')
plt.title('$y=x^2-x+2$')
plt.annotate('origin', xy=(0,0), xytext=(0,100), arrowprops={})
```

## Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

$ y=\sqrt x $

```python
x = list(range(1000))
y = [n ** 0.5 for n in x]

plt.figure(figsize=(8, 4.5))
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.title('$y=\sqrt{x}$')
```

$y = x ^ 3$

```python
x = list(range(25))
y = [n**3 for n in x]

plt.figure(figsize = (8, 4.5))
plt.plot(x, y)
plt.xlabel("$x$")
plt.ylabel("$x^3$")
plt.title("$y=x^3$")
```

For the next graph we will use itertools to all us to use a more refined set of points so we can accuratly express tan(x)

```python
import itertools
def seq(start, end, step):
    if step == 0:
        raise ValueError("step must not be 0")
    sample_count = int(abs(end - start) / step)
    return itertools.islice(itertools.count(start, step), sample_count)
```

$y=\tan(x)$

```python
import math
x = list(seq(-10, 10, 0.1))
y = [math.tan(n) for n in x]

plt.figure(figsize=(8, 4.5))
plt.plot(x, y)
plt.xlabel("$x$")
plt.ylabel("$tan(x)$")
plt.title("$y=tan(x)$")
```

$y = 2^x$

```python
x = list(range(20))
y = [2 ** n for n in x]

plt.figure(figsize=(8, 4.5))
plt.plot(x, y)
plt.xlabel("$x$")
plt.ylabel("$2^x$")
plt.title("$y=2^x$")
```

## Now let's combine the last 4 graphs into one large figure with 4 subplots.

```python
x1 = list(range(1000))
y1 = [n ** 0.5 for n in x1]

x2 = list(range(25))
y2 = [n**3 for n in x2]

x3 = list(seq(-10, 10, 0.1))
y3 = [math.tan(n) for n in x3]

x4 = list(range(20))
y4 = [2 ** n for n in x4]

plt.figure(figsize=(16, 16))

plt.subplot(221)
plt.plot(x1, y1)
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.title('$y=\sqrt{x}$')

plt.subplot(222)
plt.plot(x2, y2)
plt.xlabel("$x$")
plt.ylabel("$x^3$")
plt.title("$y=x^3$")

plt.subplot(223)
plt.plot(x3, y3)
plt.xlabel("$x$")
plt.ylabel("$tan(x)$")
plt.title("$y=tan(x)$")

plt.subplot(224)
plt.plot(x4, y4)
plt.xlabel("$x$")
plt.ylabel("$2^x$")
plt.title("$y=2^x$")

plt.suptitle("Graphs")
```

Finaly, let's combine the figures we created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

```python
x1 = list(range(5))
y1 = [n ** 0.5 for n in x1]

x2 = list(range(5))
y2 = [n**3 for n in x2]

x3 = list(range(5))
y3 = [math.tan(n) for n in x3]

x4 = list(range(5))
y4 = [2 ** n for n in x4]

plt.figure(figsize=(16, 9))
plt.plot(x1, y1, label='$\sqrt{x}$')
plt.plot(x2, y2, label='$x^3$')
plt.plot(x3, y3, label='$tan(x)$')
plt.plot(x4, y4, label='$2^x$')
plt.legend(loc='upper left')
```
