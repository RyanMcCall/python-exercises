## Use the iris database to answer the following quesitons:

```python
iris = sns.load_dataset('iris')
iris
```

### What does the distribution of petal lengths look like?

```python
sns.distplot(iris.petal_length)
```

The distribution of petal lengths appears to be bimodal, so we may need to see if there would be a way to separate the flowers into groups.

### Is there a correlation between petal length and petal width?

```python
sns.regplot(x='petal_length', y='petal_width', data=iris)
```

There does seem to be a positive correlation between petal length and petal width.

We can also see that there is definitely two groupings in the scatter plot.

### Which features would be best used to predict species?

```python
sns.pairplot(iris, hue='species')
```

So above is a pair plot of the iris database with the addition of higlighting what points come from which species. From this, we are able to to see that pedal width and pedal length pretty clear definition between the three groups. Petal length and petal Width might be the best features to predict species.

### Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. (Choose a cutoff point for what constitutes catholic)

I will use a supermajority of 75% to represent whether the province is Catholic.

```python
swiss['is_catholic'] = swiss.Catholic >= 75
swiss
```

### Does whether or not a province is Catholic influence fertility?

```python
sns.boxplot(data=swiss, y='Fertility', x='is_catholic')
plt.xlabel("\nProvince over 75% Catholic")
plt.ylabel('Fertility\n')
plt.title("Does Being a Catholic Province Influence Fertility")
```
